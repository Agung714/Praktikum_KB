```python
dftypecount=df['type'].value_counts()
plt.bar(dftypecount.index, dftypecount,color=['blue','red'])
plt.xlabel('Varian Wine')
plt.ylabel('Jumlah')
plt.title('Perbandingan Varian Wine')
plt.show()
```

Menampilkan grafik batang,

Kita dapat melihat
sebagian besar sampel menggunakan White Wine

```python

dt1 = df.groupby(df["type"]).size()
dt2 = dt1.index
plt.figure(figsize=(7.5, 15))
explode = (0.05, 0.05)
colors  = ("Red","blue")
plt.pie(dt1, labels=dt2, explode=explode, colors=colors, autopct="%0.1f%%", shadow=True, startangle=90)
# plt.pie(dt1)
plt.title("Grafik Pie Perbandingan Varian Wine")
plt.legend(['Red', 'White'])
plt.show()
```

Menamppilkan Pie Chart

Kita dapat melihat
75,4% dari sampel yang digunakan adalah white wine,
sedangkan 24,6% sisanya adalah red wine

```python
ukuran=plt.figure(figsize=(12,18))
plt.title("Grafik Heatmap")
sns.heatmap(
    df.corr(),
    annot=True
    )
ukuran.show()
```

Menampilkan grafik heatmap
Kita dapat melihat
quality memiliki korelasi tertinggi dengan alcohol senilai 0.44 dan
quality memiliki korelasi terendah dengan density senilai -0.31

dapat disimpulkan jika tingkat alcohol ditingkatkan dan density dikurangi maka quality akan meningkat begitu juga sebaliknya
